### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use `sum(axis=1)` store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.  

```python
movie_data_group = movie_data.drop(columns=['movie id','movie title','release date'])
movie_data_group = movie_data_group.groupby('release_year',as_index=True).sum()
movie_data_group['total'] = movie_data_group[movie_data_group.columns].sum(axis=1)
movie_data_group = (movie_data_group.div(movie_data_group['total'],axis=0)*100).round(2)
movie_data_group.drop(columns=['total','unknown'],inplace=True)
fig,ax = plt.subplots(figsize=(25,45))
sns.heatmap(movie_data_group,cmap="Greens")
```

### 10. See gender distribution across different genres check for the validity of the below statements - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#Merge all datasets
data_rating_user = user_data.merge(rating_data,how='inner',on='user id')
full_data = pd.DataFrame(data_rating_user.merge(movie_data,how='inner',on='movie id'))

#dividing the data into Male and Female
female_data = full_data.where(full_data.gender =='F',inplace=False).dropna()
male_data = full_data.where(full_data.gender =='M',inplace=False).dropna()
```

```python
#Calculating the % of Drama viewers among the genders
#Female %
female_drama = female_data['Drama'].sum()/female_data.iloc[:,10:-1].sum().sum()*100
#Male %
male_drama = male_data['Drama'].sum()/male_data.iloc[:,10:-1].sum().sum()*100
#Comparing the data and result
print(f"Percentage Female watch Drama: {female_drama.round(2)}%\nPercentage Male watch Drama: {male_drama.round(2)}%\nDo Female watch more Drama then Men: {female_drama > male_drama}")
```

```python
#Calculating the % of Sci-Fi viewers among the genders
#Female %
female_scifi = female_data['Sci-Fi'].sum()/female_data.iloc[:,10:-1].sum().sum()*100
#Male %
male_scifi = male_data['Sci-Fi'].sum()/male_data.iloc[:,10:-1].sum().sum()*100
#Comparing the data and result
print(f"Percentage of Female who watch Sci-Fi: {female_scifi.round(2)}%\nPercentage of Male who watch Sci-Fi: {male_scifi.round(2)}%\nDo Women watch more Sci-Fi than men: {female_scifi > male_scifi}")
```

```python
#Calculating Men watch more Romance than women
#Female %
female_romance = female_data['Romance'].sum()/female_data.iloc[:,10:-1].sum().sum()*100
#Male %
male_romance = male_data['Romance'].sum()/male_data.iloc[:,10:-1].sum().sum()*100
#Comparing the data and result
print(f"Percentage of Female who watch the genre Romance: {female_romance.round(2)}%\nPercentage of Male who watch the genre Romance: {male_romance.round(2)}%\nDo Men watch more Romance than women: {female_romance < male_romance}")
```
